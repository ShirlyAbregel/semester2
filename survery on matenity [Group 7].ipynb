## Third conclusion <a class="anchor" id="5-bullet"></a>

#### Q10: "If a woman chooses not to have a child, she is making a poor decision".
#### Q12: "A woman who chooses to not have a child is selfish".

#### Display of the results of questions:

```python
fig, axes = plt.subplots(figsize=(27, 7), ncols=2)
sns.histplot(data=Survey_df,color = "rosybrown" ,x='Q10',ax = axes[0]);
axes[0].set_xlabel(' Q10 If a woman chooses not to have a child, she is making a poor decision.' ,fontsize=18);
sns.histplot(data=Survey_df,color = "lightcoral" , x='Q12',ax = axes[1]);
axes[1].set_xlabel(' Q12 A woman who chooses to not have a child is selfish.',fontsize=18);
fig.suptitle('Presentation of questions number 10 and 12 that show the conflict between women regarding having children',fontsize=25);
```

#### It is evident that there is a strong negative match between the two questions

```python
corr_prep = Survey_df[['Q12','Q10',]] #Calculation of correlation between the Q12 and Q10
corr_prep.corr(method = 'pearson')
```

#### The pearson correlation is almost -0.8

```python
sns.heatmap(corr_prep.corr(), vmin=0.0 , vmax = 1,cmap='Reds' , annot=True);
```

```python
Survey_df=Survey_df.drop(Survey_df[Survey_df['Q10']==3].index) #Neutralizing answer number 3 "neutral" and replace values for two categories in Q10
Survey_df['Q10'].replace((4,5),'Agree',inplace=True)
Survey_df['Q10'].replace((1,2),'DisAgree',inplace=True)
```
